```python
import matplotlib.pyplot as plt
import numpy as np

years = ['2003', '2007', '2010', '2012', '2015']

sachin_runs = [1141, 1425, 1063, 315, 0]
kohli_runs = [0, 281, 995, 1026, 1383]
sehwag_runs = [1000, 1420, 1097, 844, 619]

x = np.arange(len(years))
bar_width = 0.25  

#  # Side-by-side Bar Chart 
print("Showing bar chart for Sachin, Kohli and Sehwag")

plt.style.use('seaborn-v0_8-darkgrid')  
plt.figure(figsize=(10, 6))

plt.bar(x - bar_width, sachin_runs, width=bar_width, label='Sachin', color='orange')
plt.bar(x, kohli_runs, width=bar_width, label='Kohli', color='blue')
plt.bar(x + bar_width, sehwag_runs, width=bar_width, label='Sehwag', color='green')

plt.title("Runs Comparison - 5 Years")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.xticks(x, years)
plt.legend()
plt.show()


# Horizontal Bar Chart

sachin_debut = [215, 820, 625, 1020, 870]
kohli_debut = [325, 995, 1026, 1050, 973]
sehwag_debut = [282, 800, 900, 756, 889]

total_sachin = sum(sachin_debut)
total_kohli = sum(kohli_debut)
total_sehwag = sum(sehwag_debut)

players = ['Sachin', 'Kohli', 'Sehwag']
totals = [total_sachin, total_kohli, total_sehwag]

print("Now showing total runs in first 5 years")

plt.figure(figsize=(8, 5))
plt.barh(players, totals, color=['orange', 'blue', 'green'])
plt.title("Debut 5 Years - Total Runs")
plt.xlabel("Runs")
plt.show()
```

```python
import matplotlib.pyplot as plt

os_names = ['Android', 'iOS', 'Others']
market_share = [72, 26, 2] 

# Pie Chart 
print("Displaying pie chart for mobile OS market share")

plt.figure(figsize=(6, 6))
plt.pie(market_share, labels=os_names, autopct='%1.1f%%', startangle=90, colors=['green', 'lightblue', 'grey'])
plt.title("Mobile OS Market Share (Pie Chart)")
plt.show()

# Horizontal Bar Chart
print("Displaying horizontal bar chart for same data")

plt.figure(figsize=(8, 4))
plt.barh(os_names, market_share, color=['green', 'lightblue', 'grey'])
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share (Bar Chart)")
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D", "E"]
values = [15, 25, 30, 20, 10]

plt.bar(categories, values, color='skyblue')

for i, val in enumerate(values):
    plt.text(i, val + 0.5, str(val), ha='center', fontsize=10)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Annotations")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100) 
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(x/4) 

plt.plot(x, y1, color='blue', linestyle='-', label='sin(x)')
plt.plot(x, y2, color='red', linestyle='--', label='cos(x)')
plt.plot(x, y3, color='black', linestyle=':', label='exp(x/4)')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines in One Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y = np.array([2, 4, 6, 8, 10])

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot(x, y, marker='o', color='blue')
axs[0, 0].set_title("Line Plot")

axs[0, 1].bar(x, y, color='green')
axs[0, 1].set_title("Bar Chart")

axs[1, 0].scatter(x, y, color='red', s=80)
axs[1, 0].set_title("Scatter Plot")

data = np.random.randn(100)
axs[1, 1].hist(data, bins=10, color='orange', edgecolor='black')
axs[1, 1].set_title("Histogram")

fig.suptitle("2x2 Grid of Different Plots", fontsize=14)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)

product_A = np.array([5, 7, 8, 6, 7, 9, 10, 12, 11, 9, 10, 12])
product_B = np.array([3, 4, 4, 5, 6, 6, 7, 8, 7, 6, 8, 9])
product_C = np.array([2, 3, 3, 4, 4, 5, 5, 6, 5, 4, 6, 7])

plt.stackplot(months, product_A, product_B, product_C, 
              labels=["Product A", "Product B", "Product C"], 
              colors=["orange", "skyblue", "lightgreen"])

plt.xlabel("Month")
plt.ylabel("Sales (in units)")
plt.title("Cumulative Sales of Products Over 12 Months")
plt.legend(loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    "Math": np.random.randint(50, 100, 20),
    "Science": np.random.randint(45, 95, 20),
    "English": np.random.randint(55, 98, 20)
}

df = pd.DataFrame(data)
df.plot(kind='box', color={'boxes':'yellow', 'whiskers':'black', 'medians':'red', 'caps':'black'})
plt.title("Student Exam Scores Distribution")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

companies = ["Company A", "Company B", "Company C", "Company D"]
market_share = [35, 25, 20, 20]

explode_values = [0.1 if share == max(market_share) else 0 for share in market_share]

plt.pie(market_share, labels=companies, autopct="%1.1f%%", 
        explode=explode_values, colors=["blue", "red", "green", "purple"], 
        shadow=True, startangle=90)

plt.title("Market Share of Companies")
plt.show()
```
