# 10 Different matplotlib styles

```python
import matplotlib.pyplot as plt

subjects = ['Math', 'Science', 'English', 'History', 'Art']
marks = [88, 75, 90, 70, 80]

plt.style.use('default')  
plt.plot(subjects, marks, color='yellow')
plt.title("My Marks in 5 Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
