```python
import matplotlib.pyplot as plt
import numpy as np

years = ['2003', '2007', '2010', '2012', '2015']

sachin_runs = [1141, 1425, 1063, 315, 0]
kohli_runs = [0, 281, 995, 1026, 1383]
sehwag_runs = [1000, 1420, 1097, 844, 619]

x = np.arange(len(years))
bar_width = 0.25  

#  # Side-by-side Bar Chart 
print("Showing bar chart for Sachin, Kohli and Sehwag")

plt.style.use('seaborn-v0_8-darkgrid')  
plt.figure(figsize=(10, 6))

plt.bar(x - bar_width, sachin_runs, width=bar_width, label='Sachin', color='orange')
plt.bar(x, kohli_runs, width=bar_width, label='Kohli', color='blue')
plt.bar(x + bar_width, sehwag_runs, width=bar_width, label='Sehwag', color='green')

plt.title("Runs Comparison - 5 Years")
plt.xlabel("Years")
plt.ylabel("Runs")
plt.xticks(x, years)
plt.legend()
plt.show()


# Horizontal Bar Chart

sachin_debut = [215, 820, 625, 1020, 870]
kohli_debut = [325, 995, 1026, 1050, 973]
sehwag_debut = [282, 800, 900, 756, 889]

total_sachin = sum(sachin_debut)
total_kohli = sum(kohli_debut)
total_sehwag = sum(sehwag_debut)

players = ['Sachin', 'Kohli', 'Sehwag']
totals = [total_sachin, total_kohli, total_sehwag]

print("Now showing total runs in first 5 years")

plt.figure(figsize=(8, 5))
plt.barh(players, totals, color=['orange', 'blue', 'green'])
plt.title("Debut 5 Years - Total Runs")
plt.xlabel("Runs")
plt.show()
```

```python
import matplotlib.pyplot as plt

os_names = ['Android', 'iOS', 'Others']
market_share = [72, 26, 2] 

# Pie Chart 
print("Displaying pie chart for mobile OS market share")

plt.figure(figsize=(6, 6))
plt.pie(market_share, labels=os_names, autopct='%1.1f%%', startangle=90, colors=['green', 'lightblue', 'grey'])
plt.title("Mobile OS Market Share (Pie Chart)")
plt.show()

# Horizontal Bar Chart
print("Displaying horizontal bar chart for same data")

plt.figure(figsize=(8, 4))
plt.barh(os_names, market_share, color=['green', 'lightblue', 'grey'])
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share (Bar Chart)")
plt.show()
```
